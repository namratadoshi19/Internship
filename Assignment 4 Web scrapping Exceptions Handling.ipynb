```python
time.sleep(5)
try:
    element=driver.find_element(By.XPATH,"//nav[@id='global-nav']/div/a[5]")
    element.click()
except NoSuchElementException as e:
    print("Exception Raised: ",e)
    element=driver.find_element(By.XPATH,"//a[@class='HeaderMenu-dropdown-link lh-condensed d-block no-underline position-relative py-2 Link--secondary']")
    element.click()
    print(element)
```

```python
title=[]
des=[]
count=[]
lang=[]
```

```python
product_urls=[]
start=0
end=1
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/main/div[3]/div/div[2]/article/h2/a')
    for i in url:
        product_urls.append(i.get_attribute("href"))
```

```python
for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        TT=driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/main/div/div[1]/div[1]/div/span[1]/a')
        title.append(TT.text)
    except NoSuchElementException:
        title.append("-")
        
for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        DS=driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/main/turbo-frame/div/div/div/div[2]/div[2]/div/div[1]/div/div/p')
        des.append(DS.text)
    except NoSuchElementException:
        des.append("-")
        
for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        CT=driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/main/turbo-frame/div/div/div/div[2]/div[2]/div/div/div/h2/a/span')
        count.append(CT.text)
    except NoSuchElementException:
        count.append("-")
        
for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        LG=driver.find_element(By.XPATH,'//ul[@class="list-style-none"]')
        lang.append(LG.text.replace("\n",""))
    except NoSuchElementException:
        lang.append("-")
```

```python
print(len(title),len(des),len(count),len(lang))
```

```python
GH=pd.DataFrame({"Rep_title": title,"Rep_des": des,"Cont_count": count,"Lang_used":lang})
GH
```
